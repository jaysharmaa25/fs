```python
Number = int(input("How many terms you wanna display?"))
Num1,Num2 =0,1 
Count=0
if Number<=0:
    print("Please enter the number greate than O")
elif Number==1:
    print ("Fibonacci sequence upto", Number,".")
    print (Num1) 
else:
    print ("Fibonacci sequence ")
    while Count<Number:
        print(Num1)
        Nth=Num1+Num2
        Num1=Num2
        Num2=Nth
        Count+=1
```
